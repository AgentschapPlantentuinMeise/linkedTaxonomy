## Example query wikidata to get all taxon names that have no Zoobank ID


```
select ?item ?itemLabel ?ipni ?plaziID where {
  ?item wdt:P31 wd:Q16521;
        wdt:P105 wd:Q7432;
        wdt:P961 ?ipni;
        wdt:P1992 ?plaziID.
  FILTER NOT EXISTS {?item wdt:P1746 ?zoo.}.
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}
LIMIT 2000
```
